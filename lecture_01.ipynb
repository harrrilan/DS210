## Why Rust?

* Important to learn a "systems" programming language
* Generally, compiled languages such as Rust are needed to scale to large, efficient deployments
* Can be 10x to 100x faster than equivalent Python

### And Rust is "memory safe"

Google, Microsoft, Meta have efforts underway to move their infrastructure code to Rust
and other memory safe languages from C/C++.

### Big-Oh Notation in Programming

* **What is Big-O Notation?** Describes algorithm efficiency and scalability relative to size of input, say $n$
* **Focus on Worst-Case (asymptotic) Scenario**  Helps analyze the upper bound of performance.
* **Common Complexity Classes**  O($1$) (constant time), O($\log n$), O($n$), O($n \log n$), O($n^2$) (quadratic), O($n^3$) (cubic), O($2^n$) (exponential) etc.
* **Why It Matters** Guides optimization and decision-making.
* **Practical Examples** `Vec` operations: `push()` can be O($1$), bubble sort O($n^2$), optimal `sort()`, e.g. `timsort()` is O($n \log n$).


(http://interactivepython.org/courselib/static/pythonds/AlgorithmAnalysis/BigONotation.html)
